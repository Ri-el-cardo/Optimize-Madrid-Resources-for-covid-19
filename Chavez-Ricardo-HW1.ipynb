# Distribution of the covid-19 resources in Madrid

(first lets state that this situation is hypothetical)

There is a new treatment for covid19.
The deployment of this treatment requires qualified personal physician in a hospital that satisfies certain special conditions (not every hospital is valid)

Our objective is to deploy this treatment such as the most number of people is covered given the budget.
For this we hav to take into account that different hospitals, due to its geografical position and its installations, can treat a different number of cases with the same ammount of money, and that each hospital treats a different number of cases in each different region.

In order to distribute the resources fairly, each region must have the capability to treat at least 1% of its total population.

### Variables

$x_j \equiv\ €\ spent\ in\ hospital\ \textbf{j}\ deploying\ the\ cure$

### Objective function
minimize the money spent

$min\sum_{j\in J} x_j $

### Constraints
$\sum_{j\in J} x_j A_{i,j} >= Resources_i; \forall\ \textbf{i}\in \textbf{I}$

$x_j  <= Budget_j; \forall\ \textbf{j}\in \textbf{J}$

```python
from pyomo.environ import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.core.display import Image, display, Math, Latex

%matplotlib inline

plt.style.use('seaborn-talk')
plt.style.use('seaborn-colorblind')
```

```python
url = "Data/Hospitales Madrid.csv"
data = pd.read_csv(url,index_col = 0)
display(data)
```

## b) Implement the model

```python
M = ConcreteModel()

I = list(data.iloc[:-1,:-1].index.values) 
J = list(data.iloc[:-1,:-1].columns.values)
M.I = Set(initialize=I)
M.J = Set(initialize=J)

A_data = data.iloc[:-1,:-1].stack().to_dict()
M.A = Param(M.I,M.J,initialize=A_data)

M.x = Var(M.J, within=NonNegativeReals,doc='Money spent in hospital j')

Resources_data = data.iloc[:-1,-1].to_dict()
M.Resources = Param(M.I,initialize=Resources_data,within=NonNegativeReals)

Budget_data = data.iloc[-1,:-1].to_dict()
M.Budget = Param(M.J, initialize=Budget_data,within=NonNegativeReals)
```

```python
def objective_rule(M):
    return sum(M.x[j] for j in M.J)

M.obj = Objective(rule=objective_rule, sense=minimize, doc='deployment cost')

M.constraints = ConstraintList()
  
def Min_resources(i):
    return sum(M.x[j]* M.A[i,j] for j in M.J) >= M.Resources[i]

def Budget_lim (j):
    return M.x[j] <= M.Budget[j]

for i in I:
    M.constraints.add(sum(M.x[j]* M.A[i,j] for j in M.J) >= M.Resources[i])
    
for j in J:
    M.constraints.add(M.x[j] <= M.Budget[j])
```

```python
Solver = SolverFactory('glpk')
M.dual = Suffix(direction=Suffix.IMPORT) 

Results = Solver.solve(M)
# M.x.display()
print('Total money expended ->', np.round(M.obj(),2),'€')
```

```python
results_dataframe = pd.DataFrame({"Hospital":list(data.iloc[:-1,:-1].columns.values), 
             "Money Expended (€)": [round(M.x[j].value,2) for j in J]},index = ['' for j in J])
display(results_dataframe)
```

This solution tells us that to expend the least amount of money, we need to spend that much money in each hospital.

We can see that we won't give any money to Hospital Santa Cristina. This is because the regions that this hospital covers are already covered by another hospital but for less money.

Each sensibility is corresponded with one constraint. The ones in the upper dataframe correspond with the constraint of how much people must be able to be treated in each region.

$\sum_{j\in J} x_j A_{i,j} >= Resources_i; \forall\ \textbf{i}\in \textbf{I}$

As $Resources_i = 1\% Population_i ; \forall i \in I$, each sensibility says how much should we spend if for the region $\textbf{i}$ if there were one more person to treat (100 more people in the region).

$sensibility_{Resources_i} \equiv \ the\ difference \ in \ the \ ammount\ of\ money\ that\ we\ should\ expend\ if\ the\ region\ \textbf{i}\ had\ 100\ more\ population$



For the sensibilities in the lower dataframe we can have a similar reading.
The constraint in this case is 

$x_j  <= Budget_j; \forall\ \textbf{j}\in \textbf{J}$

$sensibility_{Budget_j} \equiv \ the\ difference \ in \ the \ ammount\ of\ money\ that\ we\ should\ expend\ if\ the\ Hospital\ \textbf{j}\ had\ 1\ more\ euro\ of\ budget$

We can see that most hospitals have 0 sensibility. This is because they are not expending their whole Budget, so an increment in this would not make any difference.
We observe that the ones that have a non zero sensibility have a negative one.
This is because they have already spend all of their budget and by giving them more budget, they could cover the same regions that other hospitals are covering but without that much expense of money as they are more efficient. So, by increasing their budget the total money spend would be lower.

## d) Binary model

One more option has appear that might be more efficient in order to treat the virus.

This option is the construction of a temporary Hospital exclusively for the disease in each region.
This is interesting as the construction is very expensive, but the efficien of these new temporary hospitals is much higher than with the other hospitals.

There is also one problem with these hopsitals, to reduce the spread of the disease these specialized hospitals will only nurse the region in which they are.

These campaign hospitals have fixed cost to construct and fixed capability to attend.

### New parameters
$Ay = 100000 \equiv \ Number\ of\ cases\ that\ can\ be\ treated\ by\ a \ campaign \ hospital\ (fixed \ \forall \textbf{i} \in \textbf{I}) $

$Cy = 1000000€ \equiv \ Cost\ of\ deploying\ a\ campaign \ hospital (fixed \ \forall \textbf{i} \in \textbf{I}) $

### New varible

$y_i \equiv\ \{0,1\} wheter\ if \ a\ campaign\ hospital\ will\ be\ constructed\ in\ region\ \textbf{i} ; 0 \equiv no\ ; 1 \equiv yes$

### New objective function

$min\sum_{j\in J} x_j  + \sum_{i \in I} Cy · y_i$

### New restrictions 

$(\sum_{j\in J} x_j A_{i,j}) + y_i · Ay >= Resources_i; \forall\ \textbf{i}\in \textbf{I}$

$x_j  <= Budget_j; \forall\ \textbf{j}\in \textbf{J}$

```python
M = ConcreteModel()

I = list(data.iloc[:-1,:-1].index.values) 
J = list(data.iloc[:-1,:-1].columns.values)
M.I = Set(initialize=I)
M.J = Set(initialize=J)

A_data = data.iloc[:-1,:-1].stack().to_dict()
M.A = Param(M.I,M.J,initialize=A_data)

M.x = Var(M.J, within=NonNegativeReals,doc='Money spent in hospital j')

M.y = Var(M.I, within=Binary,doc='binary indicating the construction or not of a campaign hospital in region i')

M.Ay = Param(initialize=100000)
M.Cy = Param(initialize=1000000)

Resources_data = data.iloc[:-1,-1].to_dict()
M.Resources = Param(M.I,initialize=Resources_data,within=NonNegativeReals)

Budget_data = data.iloc[-1,:-1].to_dict()
M.Budget = Param(M.J, initialize=Budget_data,within=NonNegativeReals)
```

```python
def objective_rule(M):
    return sum(M.x[j] for j in M.J) + sum([M.y[i] * M.Cy for i in M.I])

M.obj = Objective(rule=objective_rule, sense=minimize, doc='deployment cost')

M.constraints = ConstraintList()
  
def Min_resources(i):
    return (sum(M.x[j]* M.A[i,j] for j in M.J) + M.y[i] * M.Ay)>= M.Resources[i]

def Budget_lim (j):
    return M.x[j] <= M.Budget[j]

for i in I:
    M.constraints.add((sum(M.x[j]* M.A[i,j] for j in M.J) + M.y[i] * M.Ay)>= M.Resources[i])
    
for j in J:
    M.constraints.add(M.x[j] <= M.Budget[j])
```

```python
Solver = SolverFactory('glpk')
M.dual = Suffix(direction=Suffix.IMPORT) 

Results = Solver.solve(M)
# M.x.display()
# M.y.display()
print('Total money expended ->', np.round(M.obj(),2),'€')
```

```python
results_dataframex = pd.DataFrame({"Hospital":list(data.iloc[:-1,:-1].columns.values), 
             "Money Expended (€)": [round(M.x[j].value,2) for j in J]},index = ['' for j in J])
display(results_dataframex)

results_dataframey = pd.DataFrame({"Region":list(data.iloc[:-1,:-1].index.values), 
             "Build a Hospital?": ['Yes' if round(M.y[i].value,2)==1 else 'No' for i in I]},index = ['' for i in I])
display(results_dataframey)
```

We can see that by adding the possibility of the construction of these new hospitals the total money spent has decreased from 26353786€ to 21507555€. But the new hospitals is not always the way to go, it seems that for some regions, in order to cover the total resources needed at the lowest price it is better to not build the new hospital, like, for example in Sur Metropolitano or Este Metropolitano.



## d.2) Sensitivities
The built in functionality does not return the sensitivities for this model 

```python
display(M.dual.display())
```

so we are going to calculate them iteratively (as the problem seems to be easy enough so that our computer solves it very fast).


```python
original_obj = np.copy(M.obj())
constraints_list = list(data.iloc[:-1,:-1].index.values) + list(data.iloc[:-1,:-1].columns.values)
sensitivities_binary_model = {}
```

```python
for ii in constraints_list:
    M = ConcreteModel()

    I = list(data.iloc[:-1,:-1].index.values) 
    J = list(data.iloc[:-1,:-1].columns.values)
    M.I = Set(initialize=I)
    M.J = Set(initialize=J)

    A_data = data.iloc[:-1,:-1].stack().to_dict()
    M.A = Param(M.I,M.J,initialize=A_data)

    M.x = Var(M.J, within=NonNegativeReals,doc='Money spent in hospital j')

    M.y = Var(M.I, within=Binary,doc='binary indicating the construction or not of a campaign hospital in region i')

    M.Ay = Param(initialize=100000)
    M.Cy = Param(initialize=1000000)

    Resources_data = data.iloc[:-1,-1].to_dict()
    M.Resources = Param(M.I,initialize=Resources_data,within=NonNegativeReals)

    Budget_data = data.iloc[-1,:-1].to_dict()
    M.Budget = Param(M.J, initialize=Budget_data,within=NonNegativeReals)

    def objective_rule(M):
        return sum(M.x[j] for j in M.J) + sum([M.y[i] * M.Cy for i in M.I])

    M.obj = Objective(rule=objective_rule, sense=minimize, doc='deployment cost')

    M.constraints = ConstraintList()

    def Min_resources(i):
        return (sum(M.x[j]* M.A[i,j] for j in M.J) + M.y[i] * M.Ay)>= M.Resources[i]

    def Budget_lim (j):
        return M.x[j] <= M.Budget[j]

    for i in I:
        if ii == i:
            M.constraints.add((sum(M.x[j]* M.A[i,j] for j in M.J) + M.y[i] * M.Ay)>= M.Resources[i]+1)
        else:
            M.constraints.add((sum(M.x[j]* M.A[i,j] for j in M.J) + M.y[i] * M.Ay)>= M.Resources[i])

    for j in J:
        if ii == j:
            M.constraints.add(M.x[j] <= M.Budget[j]+1)
        else:
            M.constraints.add(M.x[j] <= M.Budget[j])

    Solver = SolverFactory('glpk')
    M.dual = Suffix(direction=Suffix.IMPORT) 

    Results = Solver.solve(M)
    # M.x.display()
    # M.y.display()
    sensitivities_binary_model[ii] = M.obj() - original_obj
```

```python
sensitivities_binary_model_region = pd.DataFrame({'Region':list(sensitivities_binary_model.keys())[:len(I)],
                                                 'Population sensitivities binary model (€/100 people vaiation)': np.round(list(sensitivities_binary_model.values())[:len(I)],2)},
                                                 index = ['' for i in range(len(I))])

sensitivities_binary_model_hospital = pd.DataFrame({'Hospital':list(sensitivities_binary_model.keys())[len(I):],
                                                 'Budget sensitivities binary model (€/Budget variation)': np.round(list(sensitivities_binary_model.values())[len(I):],2)},
                                                 index = ['' for i in range(len(J))])
list(sensitivities_binary_model.values())[:len(I)]
list(sensitivities_binary_model.keys())[:len(I)]
display(sensitivities_binary_model_region)
display(sensitivities_binary_model_hospital)
```

In the second dataframe the values for the sensitivity are 0 when the hospital has not spend its whole budget and the sensitivity is negative when, if by increasing its budget, the total cost would decrease.
